### a) 

```python
import pandas as pd
import numpy as np
import jenkspy
```

```python
rdf = pd.read_csv('Heart.csv')
rdf.head(3)
```

```python
rdf.corr()
```

```python
rdf.shape
```

```python
rdf.nunique()
```

#### Replacing Null Values with Mean:  

```python
rdf.isnull().sum()
```

```python
rdf['maximum heart rate achieved'].replace(np.NaN,rdf['maximum heart rate achieved'].mean(),inplace=True)
```

```python
rdf.isnull().sum()
```

#### Binning :

```python
breaks = jenkspy.jenks_breaks(rdf['age'], nb_class=3)
print(breaks)
```

```python
rdf['age_c']=pd.cut(rdf['age'], bins=breaks,labels=['29<age<49','49<age<60','60<age<77'])
```

```python
breaks = jenkspy.jenks_breaks(rdf['serum cholestoral '], nb_class=3)
print(breaks)
```

```python
rdf['serum cholestoral _c']=pd.qcut(rdf['serum cholestoral '], q=3,labels=['126<sc<226','226<sc<290','290<sc<564'])
```

```python
breaks = jenkspy.jenks_breaks(rdf['maximum heart rate achieved'], nb_class=3)
print(breaks)
```

```python
rdf['maximum heart rate achieved_c']=pd.qcut(rdf['maximum heart rate achieved'], q=3,labels=['71<max<133','133<max<160','160<max<202'])
```

```python
rdf.drop(['age','serum cholestoral ','maximum heart rate achieved'],axis=1,inplace=True)
rdf
```

#### Binarizing Dataframe:

```python
rdf.nunique()
```

```python
rdf = pd.get_dummies(rdf,columns=['gender','fasting blood sugar','exercise induced angina','chest pain type','resting electrocardiographic','age_c','serum cholestoral _c','maximum heart rate achieved_c','target'])
rdf
```

### b) 

```python
from mlxtend.frequent_patterns import apriori
```

```python
frequent_itemsets = apriori(rdf, min_support=0.2, use_colnames=True)
frequent_itemsets
```

### c) 

```python
from mlxtend.frequent_patterns import association_rules
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

####  Stroke Doesn't Happen:

```python
display(rules[rules["consequents"].apply(lambda x: 'target_0' in str(x))])
```

#### Stroke Happens: 

```python
res=rules[rules["consequents"].apply(lambda x: 'target_1' in str(x))]
display(res)
```
