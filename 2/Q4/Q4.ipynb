```python
import pandas as pd
import numpy as np
```

## b)

```python
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('correctedDiabetes.csv')
imp = SimpleImputer(strategy='mean')
imp.fit(df)
x=pd.DataFrame(imp.transform(df),  columns=['Pregnancies','Glucose','BloodPressure','SkinThickness',
                                            'Insulin','BMI','DiabetesPedigreeFunction','Age','Outcome'])

df.update(x)
df.head()
df['Glucose'].hist()
```

## c)

### StandardScaler :

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(df)
x = scaler.transform(df)
StandardScaler_df = pd.DataFrame(x,columns=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI',
                                            'DiabetesPedigreeFunction','Age','Outcome'])
StandardScaler_df
```

### Normalize

```python
from sklearn.preprocessing import normalize
```

```python
y = normalize(df, norm='l2');
Normalized_df=pd.DataFrame(y,columns=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI',
                                            'DiabetesPedigreeFunction','Age','Outcome'])
Normalized_df
```

### MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax = MinMaxScaler()
minmax.fit(df)
z = minmax.transform(df)
MinMax_df = pd.DataFrame(z,columns=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI',
                                            'DiabetesPedigreeFunction','Age','Outcome'])
MinMax_df
```

## d)

```python
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(1, 4, sharey=True,tight_layout=True,figsize=(16,4))
axs[0].hist(StandardScaler_df['Glucose'])
axs[0].set_title("StandardScaler")
axs[1].hist(Normalized_df['Glucose'])
axs[1].set_title("Normalize")
axs[2].hist(MinMax_df['Glucose'])
axs[2].set_title("MinMaxScaler")
axs[3].hist(df['Glucose'])
axs[3].set_title("Original Data")




```
