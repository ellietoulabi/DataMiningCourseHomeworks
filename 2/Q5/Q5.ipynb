```python
import pandas as pd
import numpy as np
```

## a)

```python
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('correctedDiabetes.csv')
imp = SimpleImputer(strategy='mean')
imp.fit(df)
x=pd.DataFrame(imp.transform(df),  columns=['Pregnancies','Glucose','BloodPressure','SkinThickness',
                                            'Insulin','BMI','DiabetesPedigreeFunction','Age','Outcome'])

df.update(x)
df
```

## b)

#### IQR 

```python
# IQR Outlier

Q1 = df.quantile(0.25) 
Q3 = df.quantile(0.75) 
IQR = Q3 - Q1  
mask = (df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))

pd.set_option('display.max_rows', df.shape[0]+1)

df[mask]

```

```python
boxplot = df.boxplot(figsize=(20,20),column=['Pregnancies','Glucose','BloodPressure','SkinThickness',
                                            'Insulin','BMI','DiabetesPedigreeFunction','Age','Outcome'])
```

### c)

```python
from scipy import stats
```

```python
Z_Scored = pd.DataFrame(stats.zscore(df),columns=['Pregnancies','Glucose','BloodPressure','SkinThickness',
                                            'Insulin','BMI','DiabetesPedigreeFunction','Age','Outcome'])
z_mask = (Z_Scored>3) | (Z_Scored<-3)


print("#Outliers = ",df.shape[0]-df[~z_mask].dropna().shape[0])
print("#Not outliers = ",df[~z_mask].dropna().shape[0])
df[~z_mask].dropna().head(10)
```
