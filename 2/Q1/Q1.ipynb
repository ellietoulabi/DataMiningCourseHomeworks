## c )

```python
between50and60 = vdf[(vdf['CIRCULARITY'] > 50) & (vdf['CIRCULARITY'] < 60)]
between50and60.to_csv('CIRCULARITYbetween50and60.csv', sep='\t', encoding='utf-8')
```

## d )

```python
print('[+] Mean for RADIUS_RATIO = ', vdf['RADIUS_RATIO'].mean())
print('[+] Min  for RADIUS_RATIO = ', vdf['RADIUS_RATIO'].min() )
print('[+] Max  for RADIUS_RATIO = ', vdf['RADIUS_RATIO'].max() )
print('[+] Std  for RADIUS_RATIO = ', vdf['RADIUS_RATIO'].std() )
```

## e )

```python
vdf['ElonBin'] = 'Unknown'
vdf.loc[vdf['ELONGATEDNESS'] < 30, 'ElonBin'] ='Low'
vdf.loc[((vdf['ELONGATEDNESS'] >= 30) & (vdf['ELONGATEDNESS'] <=45)), 'ElonBin'] ='Medium'
vdf.loc[vdf['ELONGATEDNESS'] > 45, 'ElonBin'] ='High'
```

```python
vdf[['ELONGATEDNESS','ElonBin']].head()
```

## f )

```python
print('[+] Name of All Unique Classes : ' , vdf.Class.unique())
print('[+] Average Of DISTANCE_CIRCULARITY For Each Unique Class :\n',vdf.groupby('Class')['DISTANCE_CIRCULARITY'].mean())

print('\n[#] To Check Values : \n')
print('bus' ,vdf.loc[vdf.Class=='bus']['DISTANCE_CIRCULARITY'].mean())
print('opel',vdf.loc[vdf.Class=='opel']['DISTANCE_CIRCULARITY'].mean())
print('saab',vdf.loc[vdf.Class=='saab']['DISTANCE_CIRCULARITY'].mean())
print('van' ,vdf.loc[vdf.Class=='van']['DISTANCE_CIRCULARITY'].mean())
```
