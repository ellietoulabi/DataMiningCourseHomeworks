### ***Normalization:***

```python
data['Cancer'].unique()
```

```python
from scipy import stats
```

```python
df = pd.DataFrame(stats.zscore(data[cancer_data.feature_names],axis=0),
                   columns=cancer_data.feature_names)

df['Cancer']=data[['Cancer']]
df.head(3)
```

<hr style = "border-top: 3px solid #000000 ; border-radius: 3px;">

<p style =" direction:rtl;text-align:right;">
مقادیر فیچرهارا به کمک zscore نرمال می کنمی
</p>

<hr style = "border-top: 3px solid #000000 ; border-radius: 3px;">


### ***Outliers:***

```python
print(f"[+] Dataframe shape before removing outliers : {df.shape}")
masked = df[~((df[cancer_data.feature_names]>3) | (df[cancer_data.feature_names] <-3))].notna()
masked['Cancer']=True
df=df[masked].dropna()
print(f"[+] Dataframe shape after removing outliers : {df.shape}")
display(df.head(3))
```

<hr style = "border-top: 3px solid #000000 ; border-radius: 3px;">

<p style =" direction:rtl;text-align:right;">
داده های پرت را که حاصل zscore آنها بیشتر از 3 و کمتر از -3 است حدف می کنمی.
</p>

<hr style = "border-top: 3px solid #000000 ; border-radius: 3px;">


### **d)**

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Cancer',axis=1)
y = df['Cancer']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

### **e)**

```python
from sklearn.neural_network import MLPClassifier
```

```python
nn = MLPClassifier(max_iter=700).fit(X_train,y_train)
```

```python
y_pred = nn.predict(X_test)
```

### **f)**

```python
print(f"[+] Mean Accuracy = {nn.score(X_test, y_test)}")
```

### **g)**

```python
from sklearn.decomposition import PCA
import seaborn as sns
```

```python
pca = PCA(n_components=2).fit(X)
pca_X = pca.transform(X)
```

```python
sns.scatterplot(x=pca_X[:,0], y=pca_X[:,1], hue=y)
```

<hr style = "border-top: 3px solid #000000 ; border-radius: 3px;">

<p style =" direction:rtl;text-align:right;">
تبدیل pca را با 2 ویژگی اعمال کرده و نمودار را رسن می کنمی.
    
</p>

<hr style = "border-top: 3px solid #000000 ; border-radius: 3px;">


### **h)**

```python
X_train, X_test, y_train, y_test = train_test_split(pca_X, y, test_size=0.2)
```

```python
nn2 = MLPClassifier(max_iter=700).fit(X_train,y_train)
```

```python
y_pred2 = nn2.predict(X_test)
```

```python
print(f"[+] Mean Accuracy = {nn2.score(X_test, y_test)}")
```
