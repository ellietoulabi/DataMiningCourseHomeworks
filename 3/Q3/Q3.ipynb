<h3 style="text-align:left;color:#945aaf;background-color:#ffffff;font-size:48p"><strong> a) </strong></h3>

```python
df = pd.read_csv('smartphone.csv')
df.head(10)
```

```python
pd.crosstab(df.Capacity,df.Company)
```

<h3 style="text-align:left;color:#945aaf;background-color:#ffffff;font-size:48p"><strong> b) </strong></h3>

```python
pd.crosstab(index=[df['Company'],df['Weight']], columns=df['inch'], margins=True)
```

<h4 style="direction:rtl;text-align:right;font-family:Yekan, sans-serif;color:#945aaf;background-color:#ffffff;font-size:48p"><strong>شرح کد :</strong> </h4>

<p style="direction:rtl;text-align:right;">اینبار که به جای 2 مولفه 3 تا داریم ابتدا شرکت و وزن را به عنوان index ها به عنوان برچسب سطرها تعریف می کنیم و سپس مقادیر inch را به عنوان ستون تعریف می کنیم.
به این صورت برای هر کارخانه وزن های موجود آن جلوی آن آمده و اینچ ها در ستون ها مطرح شده اند.
    از groupby هم می توان استفاده کرد.
</p>


<h3 style="text-align:left;color:#945aaf;background-color:#ffffff;font-size:48p"><strong> c) </strong></h3>

```python
conTable = pd.crosstab(df.OS,df.Company)
conTable
```

```python
# general
conTable.plot(kind='bar')
```

```python
#normalized
conTable_norm = conTable.div(conTable.sum(1),axis=0)
conTable_norm.plot(kind='bar',stacked=True)
```

<h4 style="direction:rtl;text-align:right;font-family:Yekan, sans-serif;color:#945aaf;background-color:#ffffff;font-size:48p"><strong>شرح کد :</strong> </h4>

<p style="direction:rtl;text-align:right;">ابتدا جدول همسانی را رسم می کنیم. با ترسیم نمودار به تنهایی با شکل اول مواجه می شویم که برای مقایسه چندان گویا و واضح نیست.
برای حل این مشکل ابتدا نمودار را با تقسیم کردن بر مجموع نرمال کرده و سپس آن را به صورت stacked ترسیم می کنیم تا نمودار فوق ایجار شود.
آنچه مشهود است این است که سیستم عامل ios تنها در شرکت apple و windows تنها در microsoft استفاده شده است اما android را در سایر شرکت ها استفاده کرده اند و سامسونگ در این میان سهم بیشتری دارد.
</p>
