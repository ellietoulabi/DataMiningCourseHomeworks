<h3 style="text-align:left;color:#945aaf;background-color:#ffffff;font-size:48p"><strong> c) </strong></h3>

<h4 style="direction:rtl;text-align:right;font-family:Yekan, sans-serif;color:#945aaf;background-color:#ffffff;font-size:48p"><strong>پاسخ :</strong> </h4>

<p style="direction:rtl;line-height:1.7;text-align:right;">
    اگر رو متغیر داشته باشیم که هردو به یک تعداد مقدار می گیرند می توان مقدار اول متغیر اول را با مقدار اول متغیر دوم و سایر مقادیر را نیز به همین صورت با یک دیگر نظیر کرد به طوری که یک زوج را تشکیل دهند. ضریب همبستگی Pearson معیاری است برای رابطه خطی بین این دو متغیر.
این ضریب که با r نمایش داده می شوند به صورت زیر محاسبه می شود:
    
    
    r = Σᵢ((xᵢ − mean(x))(yᵢ − mean(y))) (√Σᵢ(xᵢ − mean(x))² √Σᵢ(yᵢ − mean(y))²)
    
    ⁻این ضریب میتواند هددی بین 1 و -1 باشد. 1 به این معناست که این دو متغیر به شدت با هم همبسته هستند(همبستگی مثبت) یعنی مقادیر بزرگتر اولی با مقادیر بزرگتر دومی همراه است و برعکس.-1 به این معناست که به شدت بشکل منفی با هم همبسته هستند یعنی مقادیر بیشتر اولی به مقادیر کمتر دومی منتج می شود و 0 هم یعنی هیچ همبستگی با هم ندارند 
    
</p>

<p style="direction:rtl;line-height:1.7;text-align:right;">
    p-value  در روش های آماری برای تست فرضیه استفاده می شود.
    برای سهولت در تصمیم‌گیری نسبت به نتیجه آزمون فرض آماری، شاخصی به نام «مقدار احتمال» (p-Value) ارائه می‌شود. این مقدار به محقق کمک می‌کند که بدون مراجعه به جداول توزیع‌های آماری بتواند در مورد رد یا عدم رد فرض صفر تصمیم بگیرید.
    آزمون فرض آماری، به ما این اجازه را می‌دهد که به واسطه یک نمونه تصادفی در مورد یک گزاره یا واقعیت جامعه دست به قضاوت بزنیم. یک آزمون فرض آماری از دو فرضیه تشکیل شده است. فرض اولیه را به عنوان فرض صفر (Null Hypothesis) نام‌گذاری کرده‌اند. این فرض را با علامت H0 نشان می دهند.
فرض مقابل را با HA نمایش می دهند.  ما  در تحلیل داده ها به دنبال رد فرضیه صفر و تایید فرض خلاف آن هستیم. ارزش p در واقع قدرت هریک از فرض خلاف و فرض صفر را نشان می‌دهد. p-Value عددی بین ۰ و ۱ م میباشد و به شکل زیر تفسیر می‌گردد:
    
</p>
<p style="direction:rtl;line-height:1.7;text-align:right;"> p-Value کوچک (که معمولاً کوچکتر یا مساوی عدد ۰/۰۵) نشان می‌دهد که شواهد قوی علیه فرض صفر وجود دارد و بنابراین اگر p-Value در نتایج شما برای یک آزمون کوچکتر از عدد مذکور به دست آید نشان می‌دهد که شما می‌بایست فرض صفر را رد نموده و فرض خلاف آن را قبول کنید. 
    p-Value بزرگتر (که معمولاً بزرگتر از عدد ۰/۰۵) نشان می‌دهد که شواهد به دست آمده بر علیه فرض صفر ضعیف هستند و در نتیجه شما فرض صفر را می‌پذیرید
    p-Value  بسیار نزدیک به نقطه برش (یعنی به عدد ۰/۰۵) معمولاً به عنوان نمره مرزی تلقی می‌گردد. در اینصورت p-Value  گزارش می‌شود تا خوانندگان بر اساس آن 
نتیجه‌گیری خاص خودشان را انجام دهند
    </p>


<h3 style="text-align:left;color:#945aaf;background-color:#ffffff;font-size:48p"><strong> d) </strong></h3>

```python
r, p = scipy.stats.pearsonr(df.median_house_value, df.median_income)
print('R(correlation) and P-value are :',r,', ',p)
```

</p>
